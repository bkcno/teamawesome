```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Budget = pd.read_csv('actualls_all.csv')
Budget.head()
```

```python
Budget = Budget.drop(Budget.index[[[0,1,2]]])
Budget.head()
```

```python
Budget = Budget.rename(index=str, columns={"State": "Year"})
Budget.head()
```

```python
budget_df = Budget.iloc[:,0:2]
budget_df.head()
```

```python
budget_df.reset_index(drop=True,inplace=True)
budget_df.head()
```

```python
Crime = pd.read_csv('crime_all.csv')
Crime.head()
```

```python
crime_df = Crime.iloc[:,0:2]
crime_df.head()
```

```python
crime_df = crime_df.rename(index=str, columns={"Totals": "Total_Crime"})
crime_df.head()
```

```python
budget_df = budget_df.rename(index=str, columns={"Total": "Total_Budget"})
budget_df.head()
```

```python
budget_df.head
```

```python
crime_df.head
```

```python
crime_df = crime_df.drop(crime_df.index[[[11]]])
crime_df.head
```

```python
crime_df.reset_index(inplace=True)
crime_df.head()
```

```python
budget_df.reset_index(inplace=True)
budget_df.head()
```

```python
educrime_tot = pd.merge(budget_df, crime_df, on='index')
educrime_tot.head(15)
```

```python
X = educrime_tot[['Total_Budget']]
y = educrime_tot[['Total_Crime']]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```

```python
y_train.shape
```

```python
educrime_y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# The coefficients
print('Coefficients: \n', model.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, educrime_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, educrime_y_pred))
```
