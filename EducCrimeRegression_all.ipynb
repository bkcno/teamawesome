```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Budget = pd.read_csv('actualls_all.csv')
Budget.head()
```

```python
Crime = pd.read_csv('crimes_all.csv')
Crime.head()
```

```python
Budget.reset_index(inplace=True)
Budget.head()
```

```python
Crime.reset_index(inplace=True)
Crime.head()
```

```python
educrime_df = pd.merge(Budget, Crime, on="index")
educrime_df.head
```

```python
educrime_df.head()
```

```python
#X = educrime_df[["2000_x", "2001_x","2002_x","2003_x", "2004_x","2005_x","2006_x", "2007_x","2008_x","2009_x","2010_x"]]
#y = educrime_df[["2000_y", "2001_y","2002_y","2003_y", "2004_y","2005_y","2006_y", "2007_y","2008_y","2009_y"]]
X = educrime_df[["2005_x","2006_x", "2007_x","2008_x","2009_x","2010_x"]].values.reshape(-1,1)
y = educrime_df[["2010_y"]].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create the model using LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Plot the Residuals for the Training and Testing data

plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```

```python
y_train.shape
```

```python
plt.scatter(X_train['2005_x'], model.predict(X_train))
plt.scatter(X_train['2005_x'], y_train)
```

```python
educrime_y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# The coefficients
print('Coefficients: \n', model.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, educrime_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, educrime_y_pred))
```

```python
# Plot outputs
plt.scatter(X_test['2005_x'], y_test,  color='black')
plt.plot(X_test['2005_x'], educrime_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
