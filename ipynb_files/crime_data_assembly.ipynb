```python
# Dependencies
import os
import pandas as pd
import numpy as np
```

# Crime Data

```python
# Load CSV
path = os.path.join('data', 'crime','state_totals_cleaned', '2000.csv')
crime_df = pd.read_csv(path)
crime_df.head(5)
```

```python
# Select State and Total Crime columns
crime_df = crime_df.iloc[:,[0,2]]

# Rename columns
crime_df.rename(columns={list(crime_df)[0]:'State'}, inplace=True)
```

```python
# Display crime dataframe
print('Number of rows: ', len(crime_df.iloc[:,0]))
crime_df.head(5)
```

```python
# Loop through directory and save Total Crime for each year then merge dataframes
year = 2000
directory = os.path.join('data', 'crime','state_totals_cleaned')

for file in os.listdir(directory):
    print(file)
    file_directory = 'data\\crime\\state_totals_cleaned\\'
    file_path = os.path.join(file_directory, file)
    dataframe = pd.read_csv(file_path)
    
    # Get state and Total Crime columns
    df = dataframe.iloc[:,[0,2]]
    print(df.columns.values)
    
    # If 2nd column in dataframe does not contain Total Crime 
    target = 'Violent'
    cols = [col for col in df.columns if target in col]
    print(cols)
    # Create column with Crime Total
    if cols:
        df['Crime Total'] = dataframe.iloc[:,2] + dataframe.iloc[:,7]
        df = df.iloc[:,[0,2]]
        print(df.columns.values)
    
    # Rename columns
    df.rename(columns={list(df)[0]:'State', list(df)[1]:year}, inplace=True)
    print(df.columns.values) 
    
    # Merge dataframes
    crime_df = pd.merge(crime_df, df, left_on='State', right_on='State')
    year += 1
    
crime_df = crime_df.drop(crime_df.columns[1], axis=1)         
print('Number of rows: ', len(crime_df.iloc[:,0]))
crime_df.head()
```

```python
# Display data types
crime_df.dtypes
```

```python
# Save dataframe to CSV
crime_df.to_csv('data\crime\crime_all.csv', sep=',', encoding='utf-8', index=False)
```

# Population Data
