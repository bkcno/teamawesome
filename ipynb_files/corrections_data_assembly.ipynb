```python
# Dependencies
import os
import pandas as pd
import numpy as np
```

```python
# Load CSV
path = os.path.join('data', 'corrections','09_cleaned', '2000-09f.csv')
corrections_df = pd.read_csv(path)
corrections_df.head(5)
```

```python
# Select needed columns
corrections_df = corrections_df.iloc[:,[0,16]]
```

```python
# Display corrections dataframe
corrections_df.head(5)
```

```python
# Loop through directory and save Total Corrections for each year then merge dataframes
year = 2000
directory = os.path.join('data', 'corrections','09_cleaned')

for file in os.listdir(directory):
    print(file)
    file_directory = 'data\\corrections\\09_cleaned\\'
    file_path = os.path.join(file_directory, file)
    df = pd.read_csv(file_path)
    
    df = df.iloc[:,[0,16]]
    print(df.columns.values)
    df.rename(columns={list(df)[1]:year}, inplace=True)
    
    corrections_df = pd.merge(corrections_df, df, left_on='State', right_on='State')
    year += 1
    
corrections_df = corrections_df.drop(corrections_df.columns[1], axis=1)     
    
print('Number of rows: ', len(corrections_df.iloc[:,0]))
corrections_df.head()
```

```python
corrections_df.to_csv('data\corrections\corrections_all.csv', sep=',', encoding='utf-8', index=False)
```
